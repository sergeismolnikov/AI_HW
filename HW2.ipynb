Задание 2

```python
t = np.linspace(0,10,51)

f = np.cos(t)
```

```python
title_dict = {'fontsize': 15, 'fontweight':'bold',
              'color':'Gray'}
label_dict = {'fontsize': 10, 'fontweight':'bold',
              'color':'Gray'}
```

```python
plt.axis([0.5,9.5,-2.5,2.5])

plt.plot(t,f, color = 'Green')


plt.title('График f(t)', fontdict=title_dict)
plt.ylabel('Значения t', fontdict=label_dict)
plt.xlabel('Значения f', fontdict=label_dict)

plt.show()
```

Задание 3

```python
x = np.linspace(-3,3,51)

y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows = 2, ncols=2)

# """sharey = True""" - ось для всех графиков одна

ax1, ax2, ax3, ax4 = ax.flatten()


ax1.plot(x,y1)
ax1.set_title('график y1')
ax1.set_xlim ([-5,5]) 

ax2.plot(x,y2)
ax2.set_title('график y2')

ax3.plot(x,y3)
ax3.set_title('график y3')

ax4.plot(x,y4)
ax4.set_title('график y4')

fig.set_size_inches(8,6)
plt.subplots_adjust(hspace=0.3, wspace=0.3)
```
